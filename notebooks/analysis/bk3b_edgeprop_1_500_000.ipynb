```python
df_chg = (
    df.filter(
        pl.col("year_of_sale").eq(2024) &
        pl.col("price").ge(1500000)
    ).with_columns(
        pl.col("minimum_floor").alias("min_floor"),
        pl.col("maximum_floor").alias("max_floor"),
        (99 - (pl.col("year_of_sale") - pl.col("built_year"))).alias("lease_remaining_during_sale")
    )
    .sort(pl.col("price"), descending=True)
    .select(
        "town", "year_of_sale", "month_of_sale",
        "flat_type", "road", "address",
        "min_floor", "max_floor", "price", 
        "lease_remaining_during_sale", "sqm"
    )
)
with pl.Config() as cfg:
    cfg.set_tbl_rows(30)
    display(df_chg)
```

```python
with pl.Config() as cfg:
    cfg.set_tbl_rows(30)
    display(
        df_src.join(
            df_chg, how="right", 
            left_on= ("Town", "Floor_area", "Resale_price"), 
            right_on=("town", "sqm", "price")
        )
        .sort(
            pl.col("price"), descending=True
        ).select(
            'id', 'year_of_sale', 'month_of_sale', 'town', 
            'Flat_type', 
            'road',
            'min_floor', 'max_floor', 
            'price', 'sqm'
        ).unique(
            "id"
        ).sort("price", descending=True)
    )
    
```
